```python
from PIL import Image, ImageEnhance, ImageOps
#PIL: Python Imaging Library (via Pillow), used for image processing.
#Image: Handles image loading, saving, and manipulation.
#ImageEnhance: Provides tools for enhancing image properties like brightness and contrast.
#ImageOps: Contains image processing utilities, such as flipping and grayscaling.
#os: A module to handle file system operations like creating directories and file paths.
import numpy as np # Import numpy for array conversion

import os

# The original 'image_path' was a directory, causing an IsADirectoryError.
# We need to select an actual image. Let's use `resized_images[0]` for this example.
# Since `resized_images` contains cv2 (NumPy array) images, we first convert it to a PIL Image.
image_to_augment_np = resized_images[0]
image = Image.fromarray(cv2.cvtColor(image_to_augment_np, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB for PIL

# Create a directory for augmented images
augmented_dir = "augmented_images"
#Creates the directory augmented_images
os.makedirs(augmented_dir, exist_ok=True)
 #Ensures no error is raised if the directory already exists.

# Augmentation functions
def augment_image(image, output_dir):
    # 1. Horizontal Flip
    # Flips the image horizontally (mirroring).
    flipped = ImageOps.mirror(image)
    #Saves the flipped image as flipped.jpeg in the output_dir
    flipped.save(os.path.join(output_dir, "flipped.jpeg"))

    # 2. Rotate
    rotated = image.rotate(30)
    #Rotates the image by 30 degrees (counterclockwise by default)
    rotated.save(os.path.join(output_dir, "rotated.jpeg"))

    # 3. Brightness Adjustment
    enhancer = ImageEnhance.Brightness(image)
    #Adjusts brightness by a factor of 1.5 (1.0 = original brightness).
    brighter = enhancer.enhance(1.5)
    brighter.save(os.path.join(output_dir, "brighter.jpeg"))

    # 4. Contrast Adjustment
    contrast = ImageEnhance.Contrast(image).enhance(2.0)
    #Doubles the image's contrast (2.0 = twice the original contrast).

    contrast.save(os.path.join(output_dir, "contrast.jpeg"))

    # 5. Grayscale
    grayscale = ImageOps.grayscale(image)
    grayscale.save(os.path.join(output_dir, "grayscale.jpeg"))

# Apply augmentations
augment_image(image, augmented_dir)
#Calls the augment_image function to apply all augmentations on the loaded image and save the results in the augmented_images directory.

# List the augmented files
print("Augmented images saved in:", augmented_dir)
print("Generated files:", os.listdir(augmented_dir))
```

```python
import cv2
import os

def preprocess_image(image_path, target_size=(224, 224), to_rgb=True, normalize_pixels=True):
    """
    Loads an image, resizes it, optionally converts its color format, and normalizes pixel values.

    Args:
        image_path (str): The path to the image file.
        target_size (tuple, optional): The desired (width, height) for the resized image.
                                       Defaults to (224, 224).
        to_rgb (bool, optional): If True, converts the image from BGR (OpenCV default) to RGB.
                                 Defaults to True.
        normalize_pixels (bool, optional): If True, normalizes pixel values from [0, 255] to [0, 1].
                                           Defaults to True.

    Returns:
        numpy.ndarray or None: The preprocessed image as a NumPy array, or None if the image
                                 could not be loaded.
    """
    if not os.path.exists(image_path):
        print(f"Error: Image not found at {image_path}")
        return None

    # Load image using OpenCV (loads in BGR format by default)
    img = cv2.imread(image_path)

    if img is None:
        print(f"Error: Could not load image from {image_path}")
        return None

    # Resize the image
    resized_img = cv2.resize(img, target_size)

    # Convert BGR to RGB if requested
    if to_rgb:
        processed_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)
    else:
        processed_img = resized_img

    # Normalize pixel values if requested
    if normalize_pixels:
        processed_img = processed_img.astype('float32') / 255.0

    return processed_img

print("preprocess_image() function updated with normalization option.")
```

```python
import cv2
import matplotlib.pyplot as plt

# We'll use the example_image_path from previous cells
example_image_path = favorite_team_paths[0]

# Preprocess the image without normalization for better visual clarity (0-255 range)
# We'll also ensure it's loaded as BGR initially to demonstrate conversion FROM BGR
original_bgr_img = preprocess_image(example_image_path, target_size=(200, 200), to_rgb=False, normalize_pixels=False)

if original_bgr_img is not None:
    print(f"Loaded image in BGR format with shape: {original_bgr_img.shape}")

    # Convert to Grayscale
    gray_img = cv2.cvtColor(original_bgr_img, cv2.COLOR_BGR2GRAY)

    # Convert to RGB (if not already done by preprocess_image's to_rgb=True)
    # Since preprocess_image loaded it as BGR, this explicitly converts it for display
    rgb_img = cv2.cvtColor(original_bgr_img, cv2.COLOR_BGR2RGB)

    # Convert to HSV
    hsv_img = cv2.cvtColor(original_bgr_img, cv2.COLOR_BGR2HSV)

    # Display the results
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(rgb_img) # Displaying the RGB version of the original for comparison
    plt.title('Original (RGB for display)')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(gray_img, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    # When displaying HSV, matplotlib expects RGB, so we need to convert back from HSV to RGB
    # Note: HSV values range from H [0, 179], S [0, 255], V [0, 255] in OpenCV
    hsv_display_img = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB)
    plt.imshow(hsv_display_img)
    plt.title('HSV (converted to RGB for display)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
else:
    print("Could not load or preprocess the example image for color conversions.")
```
